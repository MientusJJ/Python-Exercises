```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics

from mlxtend.plotting import plot_decision_regions
from sklearn.svm import SVC
```

# Linear SVM Classification

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200, noise=.1, random_state=42)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

# Polynomial Kernel

# Zad.

Naucz SVM z jądrem "poly" o parametrach

* degree=1, coef0=0, C=1
* degree=2, coef0=0, C=1
* degree=3, coef0=0, C=1
* degree=3, coef0=1, C=1

```python
svm = SVC(kernel='poly',degree=1, coef0=0, C=1.0)

svm.fit(X, y)


xx, yy = np.meshgrid(np.linspace(-2, 3, 500),
                     np.linspace(-1.5, 2, 500))
Z = svm.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, cmap=plt.cm.RdBu, alpha=0.5)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdBu)
plt.show()
```
